```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
digit = load_digits()
```

```python
dir(digit)
```

```python
max_depth = [i for i in range(1,30)]
c_values = [i for i in range(1,30)]
var_smoothing = [1e-10, 1e-5, 1e-1]
alpha= [1,10,30]


model_params = {

    'Random_forest':{
        'model': RandomForestClassifier(),
        'params': {
            'criterion': ['gini', 'entropy'],
            }
    },

    'SVM':{
        'model': SVC(),
        'params': {
            'C': c_values,
             'kernel': ['rbf', 'linear'],
             'gamma': ['auto', 'scale']

        }
    },

    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class = 'auto'),
        'params':{
            'C': c_values
        }
    },

    'GaussianNB':{
        'model': GaussianNB(),
        'params': {
            'priors': [None],
            'var_smoothing': var_smoothing
        }

    },

    'MultinomialNB':{
        'model': MultinomialNB(),
        'params': {
            'alpha': alpha,
            'fit_prior': [True, False]
        }
    }


}
```

```python
scores = []

for mn, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv = 5, return_train_score = False, error_score='raise')
    clf.fit(digit.data, digit.target)
    scores.append({
        'model': mn,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
result = pd.DataFrame(scores, columns = ['model', 'best_score', 'best_params'])
```
